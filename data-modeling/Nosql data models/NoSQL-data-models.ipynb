Error Correction

The 3rd query in the video failed due to an incorrect column name. The intended query from the instructor is as follows:

```python
query = "SELECT * FROM music_library WHERE year=1970 AND city='Liverpool'"
```

Additionally, suppose you want to retrieve the query results without specifying all the primary keys. In that case, you can add ALLOW FILTERING to the end of the query:

```python

query = "SELECT * FROM music_library WHERE year=1970 AND city='Liverpool' ALLOW FILTERING"
```
